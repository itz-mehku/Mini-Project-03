# **Mini Project - 03 : Subset Selection Problem**
---

### **Problem Statement No.1 :**
- Find all the subsets from a set of numbers whose sum is zero.

- **Constraint:** Subset size must be 5.
- Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

---

image.png

####**--> Solution :**

---
- Without using **Random** library:
---

```python
Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

ResultList = []
n = len(Set)

for i in range(n):
    for j in range(i+1, n):
        for k in range(j+1, n):
            for l in range(k+1, n):
                for m in range(l+1, n):
                    subset = [Set[i], Set[j], Set[k], Set[l], Set[m]]
                    if sum(subset) == 0:
                        ResultList.append(tuple(sorted(subset)))

ResultList = sorted(set(ResultList))

print("Subsets of size 5 with sum = 0 (sorted):")
for subset in ResultList:
    print(subset)
print("\nTotal subsets found:", len(ResultList))
```

---
- Using **Random** library:
---

```python
import random as r

Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
SetSize = 5
Iterations = 100000
ResultSet = set()

for _ in range(Iterations):
    subset = r.sample(Set, SetSize)
    if sum(subset) == 0:
        ResultSet.add(tuple(sorted(subset)))

SortedResult = sorted(ResultSet)

print("Subsets of size 5 with sum = 0 (randomly generated, sorted):")
for subset in SortedResult:
    print(subset)
print("\nTotal unique subsets found:", len(SortedResult))
```

####**Problem Statement No.2 :**
- Find all the subsets from a set of numbers whose sum is zero.
- **Constraint:** Subset size must be 3 to 6 only
- Set={-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}

---

####**--> Solution :**


---
- Without using **Random** library:
---

```python
Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
ResultSet = set()
n = len(Set)
# Generate combinations manually for sizes 3 to 6
for i in range(n):
    for j in range(i + 1, n):
        for k in range(j + 1, n):
            subset3 = [Set[i], Set[j], Set[k]]
            if sum(subset3) == 0:
                ResultSet.add(tuple(sorted(subset3)))

            for l in range(k + 1, n):
                subset4 = subset3 + [Set[l]]
                if sum(subset4) == 0:
                    ResultSet.add(tuple(sorted(subset4)))

                for m in range(l + 1, n):
                    subset5 = subset4 + [Set[m]]
                    if sum(subset5) == 0:
                        ResultSet.add(tuple(sorted(subset5)))

                    for o in range(m + 1, n):
                        subset6 = subset5 + [Set[o]]
                        if sum(subset6) == 0:
                            ResultSet.add(tuple(sorted(subset6)))

SortedResult = sorted(ResultSet)

print("Subsets of size 3 to 6 with sum = 0 (sorted):")
for subset in SortedResult:
    print(subset)
print("\nTotal unique subsets found:", len(SortedResult))
```

---
- Using **Random** library:
---

```python
import random as r
Set = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
ResultSet = set()
Iterations = 50000

for _ in range(Iterations):
    size = r.randint(3, 6)
    subset = r.sample(Set, size)
    if sum(subset) == 0:
        ResultSet.add(tuple(sorted(subset)))

SortedResult = sorted(ResultSet)

print("Subsets of size 3 to 6 with sum = 0 (using random):")
for subset in SortedResult:
    print(subset)
print("\nTotal unique subsets found:", len(SortedResult))
```
